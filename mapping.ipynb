```python
def find_places(searching):
    # 1. ì ‘ì† URL ë§Œë“¤ê¸° 
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(searching)
    # 2. headers ì…ë ¥í•˜ê¸°
    headers = {
    "Authorization": "KakaoAK 5f772a3ac0575909980230385e92e829"
    }
    # 3. API ìš”ì²­, ì •ë³´ ë°›ê¸°
    places = requests.get(url, headers = headers).json()['documents']
    # 4. í•„ìš”í•œ ì •ë³´ ì„ íƒí•˜ê¸°
    place = places[0]
    name = place['place_name']
    x=place['x']
    y=place['y']
    data = [name,x,y,searching]
    
    return data
```

```python
# ë°˜ë³µ ì‘ì—… ì‹¤í–‰ì‹œ ì§„í–‰ë°” í‘œì‹œí•˜ê¸°ìœ„í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ tqdm ì„¤ì¹˜í•˜ê¸°
from tqdm.notebook import tqdm
```

```python
# ì¸ìŠ¤íƒ€ê·¸ë¨ ìœ„ì¹˜ëª… ìœ„ì¹˜ì •ë³´ ê²€ìƒ‰í•˜ê¸°
# ê²€ìƒ‰ ì‹œ ì‹œê°„ì´ ë§ì´ ê±¸ë ¤, í˜„ì¬ ì‘ì—… ì§„í–‰í˜„í™©ì„ ë³¼ ìˆ˜ ìˆê²Œ, tqdm_notebook ì¶”ê°€
import time
locations_inform = []
for locations in tqdm(locations):
    try:
        data = find_places(locations)
        locations_inform.append(data)
        time.sleep(0.5)
    except:
            pass
locations_inform
```

```python
# ìœ„ì¹˜ì •ë³´ ì €ì¥í•˜ê¸°
locations_inform_df = pd.DataFrame(locations_inform)
locations_inform_df.columns = ['name_official', 'ê²½ë„', 'ìœ„ë„', 'ì¸ìŠ¤íƒ€ìœ„ì¹˜ëª…']
locations_inform_df.to_excel('./files/daegu_locations.xlsx',index=False)
```

```python
# ì¸ìŠ¤íƒ€ ê²Œì‹œëŸ‰ ë° ìœ„ì¹˜ì •ë³´ ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
locations_counts_df = pd.read_excel('./files/daegu_locations.xlsx',index_col = 0)
locations_inform_df = pd.read_excel('./files/daegu_locations.xlsx')
```

```python
# ìœ„ì¹˜ ë°ì´í„° ë³‘í•©í•˜ê¸°
location_data = pd.merge(locations_inform_df, location_counts_df,
                         how = 'inner', left_on = 'name_official', right_index=True)
location_data.head()
```

```python
location_data['name_official'].value_counts()
```

```python
# ì¥ì†Œ ì´ë¦„ ê¸°ì¤€ ë³‘í•©í•˜ê¸°
location_data = location_data.pivot_table(index=['name_official', 'ê²½ë„', 'ìœ„ë„'], values='place', aggfunc='sum')
location_data.head()
```

```python
# ì§€ë„ í‘œì‹œí•˜ê¸°
import folium
daegu = ['128.601805491082', '35.8713802646523']
map_daegu = folium.Map(location = daegu, zoom_start = 3)

for i in range(len(location_data)):
    name = location_data['name_official'][i]
    count = location_data['place'][i]
    size = int(count) * 8 #2,4,6 ë„ ìƒê³¼ì—†ìŒ
    long = float(location_data['ìœ„ë„'][i])
    lat = float(location_data['ê²½ë„'][i])
    folium.CircleMarker((long,lat),radius=size, color='red', popup=name).add_to(map_daegu)

map_daegu
```

```python
# ì§€ë„ ì €ì¥í•˜ê¸°
map_daegu.save('./files/daegu_cafe.html')
```

```python
from folium.plugins import MarkerCluster

locations = []
names = []

for i in range(len(location_data)):
    data = location_data.iloc[i]
    locations.append((float(data['ìœ„ë„']), float(data['ê²½ë„'])) )
    names.append(data['name_official'])
    
daegu = ['128.601805491082', '35.8713802646523']
map_daegu2 = folium.Map(location = daegu, zoom_start = 3)

marker_cluster = MarkerCluster(
    locations=locations,popups=names,
    name='daegu',
    overlay=True,
    control=True
)
marker_cluster.add_to(map_daegu2)
folium.LayerControl().add_to(map_daegu2)

map_daegu2
```
