```python
# ÌÅ¨Î°§ÎßÅ Í≤∞Í≥º Ï§ë Ìï¥ÏãúÌÉúÍ∑∏ Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
import pandas as pd
raw_total = pd.read_excel('./files/1_crawling_daegucafe_500_2.xlsx')
raw_total['tags'][:10] 
```

```python
# Ìï¥ÏãúÌÉúÍ∑∏ ÌÜµÌï© Ï†ÄÏû•ÌïòÍ∏∞
tags_total = []

for tags in raw_total['tags']:
    tags_list = tags[2:-2].split("', '")
    for tags in tags_list:
        tags_total.append(tags)
        
tags_total
```

```python
# ÎπàÎèÑÏàò ÏßëÍ≥ÑÌïòÍ∏∞(Counter)
from collections import Counter
tag_counts = Counter(tags_total)
```

```python
# Í∞ÄÏû• ÎßéÏù¥ ÏÇ¨Ïö©Îêú Ìï¥ÏãúÌÉúÍ∑∏ ÏÇ¥Ìé¥Î≥¥Í∏∞
tag_counts.most_common(50)
```

```python
# Îç∞Ïù¥ÌÑ∞ Ï†ïÏ†úÌïòÍ∏∞
STOPWORDS= ['', '#ÏùºÏÉÅ', '#ÏÜåÌÜµ','#Ï¢ãÏïÑÏöî','#daily','#ootd','#ÎßûÌåî','#Ï¢ãÏïÑÏöîÎ∞òÏÇ¨','#ÏÖÄÌîº','#Ï¢ãÎ∞ò','#Îç∞ÏùºÎ¶¨Î£©','#ÌåîÎ°úÏö∞','#selfie',
           '#Îç∞ÏùºÎ¶¨','#ÏÖÄÏπ¥']
tag_total_selected = []
for tag in tags_total:
    if tag not in STOPWORDS:
        tag_total_selected.append(tag)
        
tag_counts_selected = Counter(tag_total_selected)
tag_counts_selected.most_common(50)
```

```python
# ÏãúÍ∞ÅÌôî ÎùºÏù¥Î∏åÎü¨Î¶¨ Ìò∏Ï∂ú -> ÎßâÎåÄ Ï∞®Ìä∏Î°ú Ìï¥ÏãúÌÉúÍ∑∏ ÏÇ¥Ìé¥Î≥¥Í∏∞
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
import sys

if sys.platform in ['win32','win64']:
    font_name = 'malgun gothic'
elif sys.platform == "darwin":
    font_name = "AppleGothic"
    
rc('font', family = font_name)
```

```python
# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑÌïòÍ∏∞
tag_counts_df = pd.DataFrame(tag_counts_selected.most_common(30))
tag_counts_df.columns = ['tags', 'counts']
```

```python
#ÎßâÎåÄ Ï∞®Ìä∏ Í∑∏Î¶¨Í∏∞
plt.figure(figsize=(10,8))
sns.barplot(x='counts',y='tags', data = tag_counts_df)
plt.show()
```

```python
# ÏõåÎìúÌÅ¥ÎùºÏö∞Îìú ÎùºÏù¥Î∏åÎü¨Î¶¨ Î∂àÎü¨Ïò§Í∏∞
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import platform

if platform.system() == 'Windows':
    font_path = "c:/Windows/Fonts/malgun.ttf"
elif platform.system() == 'Darwin':
    font_path = '/Users/$USER/Library/Fonts/AppleGothic,ttf'
```

```python
wordcloud=WordCloud(font_path= font_path,
                    background_color='white',
                    max_words=100,
                    relative_scaling=0.3,
                    width=800,
                    height=400
                   ).generate_from_frequencies(tag_counts_selected)
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('./files/tag_wordcloud.png')
```
